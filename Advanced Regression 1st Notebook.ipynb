# Data Cleaning

```python
train.info()
```

```python
def data_clean(df):
    df = df.drop(['Alley','MiscFeature','Fence','PoolQC','FireplaceQu'],axis=1).fillna(method='ffill')
    df = pd.get_dummies(df,columns=df.loc[:,df.dtypes==object].columns,drop_first=True)
    return df
```

```python
train = data_clean(train)
test = data_clean(test)
```

```python
Svlaue = list(test.columns)
Svlaue.append('SalePrice')

train = train[Svlaue]
```

```python
x = train.drop(['SalePrice','Id'],axis=1)
y = train['SalePrice']
```

```python
from sklearn.model_selection import KFold

fold = KFold(random_state=0,shuffle=True,n_splits=7)
```

```python
for train_index, test_index in fold.split(x,y):
    x_train,x_test = x.iloc[train_index],x.iloc[test_index]
    y_train,y_test = y.iloc[train_index],y.iloc[test_index]
```

```python
# from sklearn.preprocessing import StandardScaler
# sc = StandardScaler()
# x_train = sc.fit_transform(x_train)
# x_test = sc.transform(x_test)

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)

# from sklearn.svm import SVR
# model = SVR(kernel='rbf')
# model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_test,model.predict(x_test)))#61574.32298862017
```

```python
predictions = model.predict(test.drop('Id',axis=1))
```

```python
submission = pd.DataFrame(data=predictions,index=test.index,columns=['SalePrice']).reset_index()
submission.columns = ['Id','SalePrice']
submission['Id'] = test['Id']
submission.to_csv('1st-prediction.csv',index=False)
```
